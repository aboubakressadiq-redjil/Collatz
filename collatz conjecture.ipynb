```python
import numpy as np;
import matplotlib.pyplot as plt;

def collatz_generator(n):
    while True:
        if n % 2 == 0:
            n = n / 2;
        else:
            n = 3*n+1;
        yield n;
        if n == 1:
            break;
def collatz_func(x):
    return [k for k in collatz_generator(x)];
xnums = [k for k in x if k % 2 ==0]
c = np.array(collatz_func(11))
x = np.arange(1, len(c)+1,1);
fig, ax = plt.subplots(1, 1);
plt.xticks(xnums);
plt.xlim(0, max(x))
plt.ylim(0, max(c)+0.2*max(c))
plt.xlabel('number of steps')
plt.grid()
plt.legend()
plt.scatter(x, c)
plt.plot(x, c)
plt.show()
```
